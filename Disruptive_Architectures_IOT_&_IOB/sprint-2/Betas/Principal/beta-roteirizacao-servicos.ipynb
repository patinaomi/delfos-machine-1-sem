```python
import requests

def obter_endereco_por_cep(cep):
    response = requests.get(f'https://viacep.com.br/ws/{cep}/json/')
    if response.status_code == 200:
        return response.json()
    else:
        return None

# Exemplo de uso
cep = "05763-290"  # Substitua pelo CEP desejado
endereco = obter_endereco_por_cep(cep)
print(endereco)




# Substitua 'YOUR_API_KEY' pela sua chave de API do OpenCage
API_KEY = '68ca43091a2d4763b237ed7e446ae216'
base_url = 'https://api.opencagedata.com/geocode/v1/json'

# Função para obter coordenadas de um endereço
def obter_coordenadas(endereco):
    params = {
        'q': endereco,
        'key': API_KEY,
        'language': 'pt',
        'pretty': 1
    }
    
    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Verifica se a requisição foi bem-sucedida
        data = response.json()
        
        if data['results']:
            lat = data['results'][0]['geometry']['lat']
            lng = data['results'][0]['geometry']['lng']
            return (lat, lng)
        else:
            return (None, None)  # Retorna None se não for encontrado
    except requests.exceptions.RequestException as e:
        print(f"Erro ao buscar coordenadas: {e}")
        return (None, None)

# Dados do ViaCEP
logradouro = "Rua Áurea da Gama"
bairro = "Jardim Piracuama"
localidade = "São Paulo"
uf = "SP"

# Formatar endereço
endereco_formatado = f"{logradouro}, {bairro}, {localidade}, {uf}, Brasil"

# Obter coordenadas
coordenadas = obter_coordenadas(endereco_formatado)
print(coordenadas)
```
